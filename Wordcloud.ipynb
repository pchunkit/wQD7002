```python
import string
import sys,csv,re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np # linear algebra
import time 

sns.set_style('darkgrid')
```

```python
data = pd.read_csv('Tweets_Result2020.csv')  
data21 = pd.read_csv('Tweets_Result2021.csv')  

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
df = data21[["Text_lemmatized", "sentiment"]]
df = df[df['sentiment'] == "Negative"]

text = " ".join(review for review in df['Text_lemmatized'])
mask = np.array(Image.open('Twitter.jpeg'))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["amp","la","u","im","dont","pkp_magic","harish_pkp"])

# Generate a word cloud image
 
wordcloud = WordCloud(colormap='Reds',max_font_size=50,max_words=100,stopwords=stopwords,background_color="white",mask=mask).generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.savefig("twitter2020.png", format="png")
plt.show()
```
